```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x = df['TV'], y = df['sales'])
sns.scatterplot(x = df['radio'], y = df['sales'])
sns.scatterplot(x = df['newspaper'], y = df['sales'])
```

```python
#compare each X with y
sns.pairplot(df)
```

```python
#drop y from the features
X = df.drop('sales', axis = 1)
```

```python
X
```

```python
y = df['sales']
```

```python
y
```

Now we have to split the train and test data set. test_size = 30 means 30% of the data is left for test. 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.30, random_state = 101)
```

```python
#total number of rows in the datapoint
len(df)
```

```python
#the train/test split is done in shuffled order
X_train
```

```python
y_train
```

```python
X_test
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
# y will be predicted from this test X
model.predict(X_test)
```

# Evaluation of Regression

```python
X_test.head()
```

```python
y_test.head()
```

```python
test_prediction = model.predict(X_test)
```

```python
test_prediction
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
df['sales'].mean()
```

```python
sns.histplot(data=df, x = 'sales')
```

```python
#mean absolute error
mean_absolute_error(y_test, test_prediction )
```

Here avarage true sales value is 14.0225 and mean absolute error is 1.21

```python
mean_squared_error(y_test, test_prediction)
```

```python
#but this is a squared value. So the square root is. This is generally standard deviation of the prediction or RMSE
np.sqrt(mean_squared_error(y_test, test_prediction))
```

# Evaluating Residuals

```python
test_residuals = y_test - test_prediction
```

```python
#positive values are over estimated and negative values are under estimated
test_residuals
```

```python
#there is no clear pattern
sns.scatterplot(x = y_test, y = test_residuals)
plt.axhline(y=0, color = 'red', ls = '--')
```

```python
sns.displot(test_residuals, bins = 25, kde=True)
```

```python
import scipy as sp
```

```python
#create a figure and axis to plot on
#staright line represents normal distribution
#curved and diverged pattern represent skewed distribution
fig, ax = plt.subplots(figsize =(6,8), dpi=100)
_=sp.stats.probplot(test_residuals, plot=ax)
```

# Model Deployment

```python
final_model = LinearRegression()
```

```python
final_model.fit(X,y)
```

```python
#check beta coefficients (B2,B1,B0)
final_model.coef_
```

```python
# 0.04576465 is the coefficient for TV, 0.18853002 for radio and -0.00103749 for newspaper
X.head()
```

The more positive the coefficients the more strong correlation. i.e. if we keep fixed radio and newspaper and just increase expense on TV in 1 unit then the sales will increase 0.04576465 unit.  

```python
y_hat = final_model.predict(X)
```

```python
#true data are blue and predicted data are red
fig,axes = plt.subplots(nrows = 1, ncols = 3, figsize = (16,6))
axes[0].plot(df['TV'], df['sales'], 'o')
axes[0].plot(df['TV'], y_hat, 'o', color = 'red')
axes[0].set_ylabel('sales')
axes[0].set_title('TV spend')

axes[1].plot(df['radio'], df['sales'], 'o')
axes[1].plot(df['radio'], y_hat, 'o', color = 'red')
axes[1].set_ylabel('sales')
axes[1].set_title('Radio spend')

axes[2].plot(df['newspaper'], df['sales'], 'o')
axes[2].plot(df['newspaper'], y_hat, 'o', color = 'red')
axes[2].set_ylabel('sales')
axes[2].set_title('Newspaper spend')
```
